````python
from dotenv import load_dotenv
import os
import psycopg2

load_dotenv()

db_user = os.getenv("ENV_DB_USER")
db_password = os.getenv("ENV_DB_PASSWORD")
db_host = os.getenv("ENV_DB_HOST")
db_port = os.getenv("ENV_DB_PORT")
db_name = os.getenv("ENV_DB_NAME")

conn = psycopg2.connect(
    dbname=db_name,
    user=db_user,
    password=db_password,
    host=db_host,
    port=db_port
)

PRECOS_PIZZA = {
    1: 50.00, 
    2: 70.00, 
    3: 90.00 
}

PRECOS_RECHEIOS = {
    1: 10.00,
    2: 12.00,
    3: 8.00,
    4: 15.00,
    5: 14.00,
    6: 13.00,
    7: 9.00
}

PRECOS_COMPLEMENTOS = {
    1: 10.00,
    2: 5.00,
    3: 15.00,
    4: 7.00,
    5: 8.00,
    6: 9.00,
    7: 6.00
}

order_list = {}
id_cliente = ""
id_employee = ""
order_delivered = False

def get_catalog():
    print("Catálogo de Produtos:")
    print("1 - Pizza - pequena - R$ 50,00")
    print("2 - Pizza - média - R$ 70,00")
    print("3 - Pizza - grande - R$ 90,00")

def create_order(order_list):
    order = ""
    order_list["Id_Cliente"] = id_cliente
    while order not in [1, 2, 3]:
        get_catalog()
        order = int(input("Faça seu pedido: "))
        order_list["Status"] = "Pedido pendente"
    if order == 1:
        print("Você pediu uma pizza pequena.")
        order_list["Pizza"] = "Pequena"
        order_list["Pizza_ID"] = 1
    elif order == 2:
        print("Você pediu uma pizza média.")
        order_list["Pizza"] = "Média"
        order_list["Pizza_ID"] = 2
    elif order == 3:
        print("Você pediu uma pizza grande.")
        order_list["Pizza"] = "Grande"
        order_list["Pizza_ID"] = 3
    else:
        print("Produto não encontrado. Por favor, escolha um item do catálogo.")


def get_filling_catalog():
    print("Catálogo de Recheios:")
    print("1 - Calabresa - R$ 10,00")
    print("2 - Frango - R$ 12,00")
    print("3 - Queijo - R$ 8,00")
    print("4 - Pepperoni - R$ 15,00")
    print("5 - Bacon - R$ 14,00")
    print("6 - Atum - R$ 13,00")
    print("7 - Legumes - R$ 9,00")
    print("8 - Remover sabor escolhido")
    print("9 - Passar para escolha de complementos")

def get_complement_catalog():
    print("Catálogo de Complementos:")
    print("1 - Borda Recheada - R$ 10,00")
    print("2 - Refrigerante - R$ 5,00")
    print("3 - Sobremesa - R$ 15,00")
    print("4 - Molho Extra - R$ 7,00")
    print("5 - Queijo Extra - R$ 8,00")
    print("6 - Calabresa Extra - R$ 9,00")
    print("7 - Azeitona Extra - R$ 6,00")
    print("8 - Remover complemento escolhido")
    print("9 - Passar para finalização do pedido")

def pickup_or_delivery():
    order = ""
    while order not in [1, 2]:
        print("Escolha o tipo de pedido:")
        print("1 - Retirada no local 🍕")
        print("2 - Entrega 🛵")
        order = int(input("Digite sua escolha: "))
    if order == 1:
        pickup_order()
    elif order == 2:
        delivery_order()
    else:
        print("Opção inválida. Por favor, escolha 1 para Retirada ou 2 para Entrega.")

def pickup_order():
    order_list["Status"] = "Pedido pronto para retirada"
    print("Pedido pronto para retirada. Obrigado por escolher nossa pizzaria!")

def delivery_order():
    order_list["Status"] = "Pedido a caminho"
    print("Pedido a caminho. Obrigado por escolher nossa pizzaria!")

def finalize_order(order_list):
    print("Pedido finalizado:", order_list)
    print("Obrigado por escolher nossa pizzaria!")


def remove_filling(order_list):
    if "Recheio" in order_list and order_list["Recheio"]:
        print("Recheios atuais:", order_list["Recheio"])
        flavor_to_remove = input("Digite o nome do recheio que deseja remover: ")
        if flavor_to_remove in order_list["Recheio"]:
            order_list["Recheio"].remove(flavor_to_remove)
            print(f"{flavor_to_remove} removido do pedido.")
        else:
            print(f"{flavor_to_remove} não encontrado no pedido.")
    else:
        print("Nenhum recheio para remover.")

def remove_complement(order_list):
    if "Complementos" in order_list and order_list["Complementos"]:
        print("Complementos atuais:", order_list["Complementos"])
        complement_to_remove = input("Digite o nome do complemento que deseja remover: ")
        if complement_to_remove in order_list["Complementos"]:
            order_list["Complementos"].remove(complement_to_remove)
            print(f"{complement_to_remove} removido do pedido.")
        else:
            print(f"{complement_to_remove} não encontrado no pedido.")
    else:
        print("Nenhum complemento para remover.")

def add_filling(order_list):
    get_filling_catalog()
    if "Recheio" not in order_list:
        order_list["Recheio"] = []
        order_list["Recheio_IDs"] = []
    while True:
        prompt = "Escolha mais um sabor: " if order_list["Recheio"] else "Escolha um sabor: "
        order = int(input(prompt))

        if order == 9:
            if not order_list["Recheio"]:
                print("Você deve escolher pelo menos um recheio antes de finalizar.")
                continue
            print("Finalizando escolha de recheios.")
            break

        if len(order_list["Recheio"]) >= 3 and order in [1, 2, 3, 4, 5, 6, 7]:
            print("Você já escolheu o número máximo de recheios (3).")
            break

        if order == 1:
            print("Você adicionou calabresa.")
            order_list["Recheio"].append("Calabresa")
            order_list["Recheio_IDs"].append(1)
        elif order == 2:
            print("Você adicionou frango.")
            order_list["Recheio"].append("Frango")
            order_list["Recheio_IDs"].append(2)
        elif order == 3:
            print("Você adicionou queijo.")
            order_list["Recheio"].append("Queijo")
            order_list["Recheio_IDs"].append(3)
        elif order == 4:
            print("Você adicionou pepperoni.")
            order_list["Recheio"].append("Pepperoni")
            order_list["Recheio_IDs"].append(4)
        elif order == 5:
            print("Você adicionou bacon.")
            order_list["Recheio"].append("Bacon")
            order_list["Recheio_IDs"].append(5)
        elif order == 6:
            print("Você adicionou atum.")
            order_list["Recheio"].append("Atum")
            order_list["Recheio_IDs"].append(6)
        elif order == 7:
            print("Você adicionou legumes.")
            order_list["Recheio"].append("Legumes")
            order_list["Recheio_IDs"].append(7)
        elif order == 8:
            print("Removendo sabor")
            remove_filling(order_list)
        else:
            print("Recheio não encontrado. Por favor, escolha um item do catálogo.")
            continue

        print("Recheios atuais:", order_list["Recheio"])


def add_complement(order_list):
    get_complement_catalog()
    order = ""
    if "Complementos" not in order_list:
        order_list["Complementos"] = []
        order_list["Complementos_IDs"] = []
    while order != 9:
        prompt = "Escolha mais um complemento: " if order_list["Complementos"] else "Escolha um complemento: "
        order = int(input(prompt))
        if order == 1:
            print("Você adicionou borda recheada.")
            order_list["Borda"] = "Recheada"
            order_list["Complementos"].append("Borda Recheada")
            order_list["Complementos_IDs"].append(1)
        elif order == 2:
            print("Você adicionou refrigerante.")
            order_list["Bebida"] = "Refrigerante"
            order_list["Complementos"].append("Refrigerante")
            order_list["Complementos_IDs"].append(2)
        elif order == 3:
            print("Você adicionou sobremesa.")
            order_list["Sobremesa"] = "Sim"
            order_list["Complementos"].append("Sobremesa")
            order_list["Complementos_IDs"].append(3)
        elif order == 4:
            print("Você adicionou molho extra.")
            order_list["Molho"] = "Extra"
            order_list["Complementos"].append("Molho Extra")
            order_list["Complementos_IDs"].append(4)
        elif order == 5:
            print("Você adicionou queijo extra.")
            order_list["Queijo"] = "Extra"
            order_list["Complementos"].append("Queijo Extra")
            order_list["Complementos_IDs"].append(5)
        elif order == 6:
            print("Você adicionou calabresa extra.")
            order_list["Calabresa"] = "Extra"
            order_list["Complementos"].append("Calabresa Extra")
            order_list["Complementos_IDs"].append(6)
        elif order == 7:
            print("Você adicionou azeitona extra.")
            order_list["Azeitona"] = "Extra"
            order_list["Complementos"].append("Azeitona Extra")
            order_list["Complementos_IDs"].append(7)
        elif order == 8:
            remove_complement(order_list)
        elif order == 9:
            print("Finalizando escolha de complementos.")
        else:
            print("Complemento não encontrado. Por favor, escolha um item do catálogo.")


def update_order(order_list):
    order = int(input("Deseja atualizar sabor ou complemento? \n 1 - Sabor \n 2 - Complemento \n 3 - Finalizar atualização \n"))
    while order != 3:
        if order == 1:
            print("Pedido atual:", order_list)
            add_filling(order_list)
        elif order == 2:
            remove_complement(order_list)
            add_complement(order_list)
        else:
            print("Opção inválida. Por favor, escolha 1, 2 ou 3.")
        order = int(input("Deseja atualizar sabor ou complemento? \n 1 - Sabor \n 2 - Complemento \n 3 - Finalizar atualização \n"))

def checkout_order(order_list):
    print("Seu pedido:", order_list)
    total = calculate_order_total(order_list)
    print(f"\n💰 Total do Pedido: R$ {total:.2f}\n")
    order = int(input("Deseja finalizar o pedido? \n 1 - Sim \n 2 - Não, quero atualizar meu pedido \n"))
    if order == 1:
        finalize_order(order_list)
        pickup_or_delivery()
    elif order == 2:
        update_order(order_list)
        checkout_order(order_list)
    else:
        print("Opção inválida. Por favor, escolha 1 para Sim ou 2 para Não.")
        checkout_order(order_list)

def calculate_order_total(order_list):
    total = 0
    if "Pizza_ID" in order_list:
        total += PRECOS_PIZZA[order_list["Pizza_ID"]]
    
    if "Recheio_IDs" in order_list:
        for recheio_id in order_list["Recheio_IDs"]:
            total += PRECOS_RECHEIOS[recheio_id]
    
    if "Complementos_IDs" in order_list:
        for complemento_id in order_list["Complementos_IDs"]:
            total += PRECOS_COMPLEMENTOS[complemento_id]
    
    return total

def collect_customer_data(order_list):
    print("\n===== DADOS DO CLIENTE =====")
    order_list["Nome_Cliente"] = input("Digite seu nome: ")
    order_list["Telefone"] = input("Digite seu telefone: ")
    order_list["Email"] = input("Digite seu email: ")
    
    if order_list["Status"] == "Pedido a caminho":
        order_list["Endereco"] = input("Digite seu endereço completo para entrega: ")
    
    print("\nDados coletados com sucesso!")

def search_order_db(login_result):
    order_id = int(input("Digite o ID do pedido para buscar: "))
    with conn.cursor() as cur:
        if login_result["type"] == "client":
            cur.execute("SELECT id, pizza, status, recheios, complementos, nome_cliente, telefone, email, endereco FROM orders WHERE id = %s AND id_cliente = %s;", (order_id, login_result["id"]))
        else:
            cur.execute("SELECT id, pizza, status, recheios, complementos, nome_cliente, telefone, email, endereco FROM orders WHERE id = %s;", (order_id,))
        order = cur.fetchone()
        if order:
            print("\n===== PEDIDO ENCONTRADO =====")
            print(f"ID: {order[0]}")
            print(f"Pizza: {order[1]}")
            print(f"Status: {order[2]}")
            print(f"Recheios: {order[3]}")
            print(f"Complementos: {order[4]}")
            print(f"Cliente: {order[5]}")
            print(f"Telefone: {order[6]}")
            print(f"Email: {order[7]}")
            if order[8]:
                print(f"Endereço: {order[8]}")
            return order_id
        else:
            print("Pedido não encontrado.")
            return None


def edit_order_db(login_result):
    order_id = search_order_db(login_result)
    if order_id is None:
        return
    
    print("\nO que deseja atualizar?")
    if login_result["type"] == "employee":
        print("1 - Status do pedido")
    print("2 - Dados do cliente")
    print("3 - Cancelar")
    
    opcao = input("Digite sua escolha: ")
    
    if opcao == "1" and login_result["type"] == "employee":
        print("\nNovos status disponíveis:")
        print("1 - Pedido pendente")
        print("2 - Pedido pronto para retirada")
        print("3 - Pedido a caminho")
        print("4 - Pedido entregue")
        
        status_opcao = input("Escolha o novo status: ")
        
        status_map = {
            "1": "Pedido pendente",
            "2": "Pedido pronto para retirada",
            "3": "Pedido a caminho",
            "4": "Pedido entregue"
        }
        
        if status_opcao in status_map:
            new_status = status_map[status_opcao]
            with conn:
                with conn.cursor() as cur:
                    cur.execute("UPDATE orders SET status = %s WHERE id = %s;", (new_status, order_id))
            print(f"Status atualizado para: {new_status}")
        else:
            print("Opção inválida.")
    
    elif opcao == "2":
        print("\nQual dado do cliente deseja atualizar?")
        print("1 - Nome")
        print("2 - Telefone")
        print("3 - Email")
        print("4 - Endereço")
        
        dado_opcao = input("Digite sua escolha: ")
        
        if dado_opcao == "1":
            novo_nome = input("Digite o novo nome: ")
            with conn:
                with conn.cursor() as cur:
                    cur.execute("UPDATE orders SET nome_cliente = %s WHERE id = %s;", (novo_nome, order_id))
            print("Nome atualizado!")
        
        elif dado_opcao == "2":
            novo_telefone = input("Digite o novo telefone: ")
            with conn:
                with conn.cursor() as cur:
                    cur.execute("UPDATE orders SET telefone = %s WHERE id = %s;", (novo_telefone, order_id))
            print("Telefone atualizado!")
        
        elif dado_opcao == "3":
            novo_email = input("Digite o novo email: ")
            with conn:
                with conn.cursor() as cur:
                    cur.execute("UPDATE orders SET email = %s WHERE id = %s;", (novo_email, order_id))
            print("Email atualizado!")
        
        elif dado_opcao == "4":
            novo_endereco = input("Digite o novo endereço: ")
            with conn:
                with conn.cursor() as cur:
                    cur.execute("UPDATE orders SET endereco = %s WHERE id = %s;", (novo_endereco, order_id))
            print("Endereço atualizado!")
        
        else:
            print("Opção inválida.")
    
    elif opcao == "3":
        print("Operação cancelada.")
    else:
        print("Opção inválida.")


def save_order(order_list):
    total = calculate_order_total(order_list)
    with conn:
        with conn.cursor() as cur:
            cur.execute("""
                INSERT INTO orders (pizza, status, recheios, complementos, nome_cliente, telefone, email, endereco, total, id_cliente)
                VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
                RETURNING id;
            """, (
                order_list.get("Pizza"),
                order_list.get("Status"),
                order_list.get("Recheio", []),
                order_list.get("Complementos", []),
                order_list.get("Nome_Cliente"),
                order_list.get("Telefone"),
                order_list.get("Email"),
                order_list.get("Endereco"),
                total,
                id_cliente
            ))
            order_id = cur.fetchone()[0]
            print(f"\n✅ Pedido salvo com ID: {order_id}")
            print(f"💰 Total: R$ {total:.2f}\n")    
    order_list.clear()


def list_orders_db(login_result):
    with conn.cursor() as cur:
        if login_result["type"] == "client":
            cur.execute("SELECT id, pizza, status, nome_cliente, telefone, total FROM orders WHERE id_cliente = %s;", (login_result["id"],))
        else:
            cur.execute("SELECT id, pizza, status, nome_cliente, telefone, total FROM orders;")
        orders = cur.fetchall()
        if orders:
            print("\n===== LISTA DE PEDIDOS =====")
            for row in orders:
                print(f"ID: {row[0]} | Pizza: {row[1]} | Status: {row[2]} | Cliente: {row[3]} | Tel: {row[4]} | Total: R$ {row[5]:.2f}")
        else:
            print("Nenhum pedido registrado.")


def delete_order_db():
    order_id = int(input("Digite o ID do pedido para remover: "))
    with conn:
        with conn.cursor() as cur:
            cur.execute("DELETE FROM orders WHERE id = %s;", (order_id,))
            print("Pedido removido.")


def cancel_order_db(login_result):
    order_id = int(input("Digite o ID do pedido para cancelar: "))
    with conn:
        with conn.cursor() as cur:
            if login_result["type"] == "client":
                cur.execute("UPDATE orders SET status = 'Cancelado' WHERE id = %s AND id_cliente = %s;", (order_id, login_result["id"]))
                print("Pedido cancelado.")
                return 
            else:
                cur.execute("UPDATE orders SET status = 'Cancelado' WHERE id = %s;", (order_id,))
            print("Pedido cancelado.")
            return 
        print("Pedido não encontrado ou você não tem permissão para cancelar este pedido.")


def user_cadaster(): 
    print("\n===== CADASTRO DE USUÁRIO =====")
    name = input("Digite seu nome: ")
    email = input("Digite seu email: ")
    password = input("Digite sua senha: ")
    client = input("Você é cliente? (s/n): ")
    boolean_client = client.lower() == 's'
    
    if boolean_client:
        with conn:
            with conn.cursor() as cur:
                cur.execute("""
                    INSERT INTO users_client (nome, email, senha)
                    VALUES (%s, %s, %s);
                """, (name, email, password))
        print("Usuário cadastrado com sucesso!")
        return

    with conn:
        with conn.cursor() as cur:
            cur.execute("""
                INSERT INTO users_employee (nome, email, senha)
                VALUES (%s, %s, %s);
            """, (name, email, password))
    print("Usuário cadastrado com sucesso!")


def login():
    print("\n===== LOGIN =====")
    hasLogin = input("Você já tem uma conta? (s/n): ")
    if hasLogin.lower() == 'n':
        user_cadaster()
    user_type = input("Você é cliente ou funcionário? (c/f): ")
    email = input("Digite seu email: ")
    password = input("Digite sua senha: ")
    boolean_client = user_type.lower() == 'c'
    
    if boolean_client:
        with conn.cursor() as cur:
            cur.execute("SELECT id FROM users_client WHERE email = %s AND senha = %s;", (email, password))
            client = cur.fetchone()
            
            if client:
                print("Login de cliente bem-sucedido!")
                person = {"id": client[0], "type": "client"}
                return person
    else:
        with conn.cursor() as cur:
            cur.execute("SELECT id FROM users_employee WHERE email = %s AND senha = %s;", (email, password))
            employee = cur.fetchone()    
            if employee:
                print("Login de funcionário bem-sucedido!")
                person = {"id": employee[0], "type": "employee"}
                return person
        
    print("Email ou senha incorretos.")
    return None



def menu():
    login_result = None
    while True:
        print("\n====== Pizzaria Leo ======")
        if login_result is None:
            print("1. Fazer Login")
            print("2. Sair")
            opcao = input("Escolha uma opção: ")
            if opcao == "1":
                login_result = login()
                if login_result is None:
                    continue
            elif opcao == "2":
                print("Saindo do programa...")
                break
            else:
                print("Opção inválida! Digite novamente.")
                continue
        if login_result is not None:
            global id_cliente, id_employee
            if login_result["type"] == "client":
                id_cliente = login_result["id"]
            else:
                id_employee = login_result["id"]
        if login_result["type"] == "employee":
            print("2. Listar Pedidos")
            print("3. Buscar Pedido")
            print("4. Editar Pedido")
            print("5. Remover Pedido")
            print("6. Sair")
        else:    
            print("1. Fazer Pedido")
            print("2. Listar Pedidos")
            print("3. Buscar Pedido")
            print("4. Editar Pedido")
            print("7. Cancelar Pedido")
            print("6. Sair")

        opcao = input("Escolha uma opção: ")

        match opcao:
            case "1":
                create_order(order_list)
                add_filling(order_list)
                add_complement(order_list)
                checkout_order(order_list)
                collect_customer_data(order_list)
                save_order(order_list)
            case "2":
                list_orders_db(login_result)
            case "3":
                search_order_db(login_result)
            case "4":
                edit_order_db(login_result)
            case "5":
                delete_order_db()
            case "6":
                print("Saindo do programa...")
                break
            case "7":
                cancel_order_db(login_result)
            case _:
                print("Opção inválida! Digite novamente.")

menu()
````
